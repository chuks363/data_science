## ðŸ”¹ 6. Heatmaps

### âœ… Heatmap Example

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulate temperature data for 7 days over 10 weeks
np.random.seed(0)
temperatures = np.random.uniform(low=20, high=35, size=(7, 10))

plt.imshow(temperatures, cmap='hot', interpolation='nearest')
plt.colorbar(label="Temperature (Â°C)")
# Label axes
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
plt.yticks(ticks=range(7), labels=days)
plt.xticks(ticks=range(10), labels=[f"Week {i+1}" for i in range(10)], rotation=45)

plt.title("Average Daily Temperatures Over 10 Weeks")
plt.xlabel("Week")
plt.ylabel("Day")
plt.tight_layout()
plt.show()

```

````python
import matplotlib.pyplot as plt
import numpy as np

# Simulate temperature data for 7 days over 10 weeks
np.random.seed(0)
temperatures = np.random.uniform(low=20, high=35, size=(7, 10))

plt.imshow(temperatures, cmap='hot', interpolation='nearest')
plt.colorbar(label="Temperature (Â°C)")
# Label axes
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
plt.yticks(ticks=range(7), labels=days)
plt.xticks(ticks=range(10), labels=[f"Week {i+1}" for i in range(10)], rotation=45)

plt.title("Average Daily Temperatures Over 10 Weeks")
plt.xlabel("Week")
plt.ylabel("Day")
plt.tight_layout()
plt.show()
````

## ðŸ”¹ 7. Histogram

## ðŸ”¹ 9. Subplots (Multiple Plots)

### âœ… Using `subplot()`

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
product_A_sales = [150, 200, 250, 220, 300]
product_B_sales = [100, 180, 230, 210, 280]

plt.figure(figsize=(10, 4))
# First subplot: Product A
plt.subplot(1, 2, 1)
plt.plot(months, product_A_sales, marker='o', color='blue')
plt.title("Product A Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
# Second subplot: Product B
plt.subplot(1, 2, 2)
plt.plot(months, product_B_sales, marker='s', color='green')
plt.title("Product B Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)

plt.tight_layout()
plt.show()
```

````python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
product_A_sales = [150, 200, 250, 220, 300]
product_B_sales = [100, 180, 230, 210, 280]

plt.figure(figsize=(10, 4))
# First subplot: Product A
plt.subplot(1, 2, 1)
plt.plot(months, product_A_sales, marker='o', color='blue')
plt.title("Product A Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
# Second subplot: Product B
plt.subplot(1, 2, 2)
plt.plot(months, product_B_sales, marker='s', color='green')
plt.title("Product B Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)

plt.tight_layout()
plt.show()
````

## ðŸ”¹ 10. Integrating with Pandas

### âœ… Line Plot from DataFrame

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'Year': [2018, 2019, 2020, 2021],
    'Sales A': [200, 220, 250, 300],
    'Sales B': [180, 210, 240, 270]
})

df.plot(x='Year', y=['Sales A', 'Sales B'], kind='line', marker='o')
plt.title("Yearly Sales Comparison")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

````python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'Year': [2018, 2019, 2020, 2021],
    'Sales A': [200, 220, 250, 300],
    'Sales B': [180, 210, 240, 270]
})

df.plot(x='Year', y=['Sales A', 'Sales B'], kind='line', marker='o')
plt.title("Yearly Sales Comparison")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
````

## ðŸ”¹ 11. Saving Figures

### âœ… Save to File

```python
plt.savefig("my_plot.png", dpi=300)
```

````python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'Year': [2018, 2019, 2020, 2021],
    'Sales A': [200, 220, 250, 300],
    'Sales B': [180, 210, 240, 270]
})

df.plot(x='Year', y=['Sales A', 'Sales B'], kind='line', marker='o')
plt.title("Yearly Sales Comparison")
plt.ylabel("Sales")
plt.grid(True)
plt.savefig("my_plot.png", dpi=300)
plt.show()
````
