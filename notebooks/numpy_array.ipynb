#### 2.2 Indexing and Slicing
##### 1D Arrays (Like Python Lists):

````python
arr = np.array([10, 20, 30, 40])
print(arr[0])      # 10  
print(arr[1:3])     # [20, 30]
````

##### 2D Arrays (Row, Column Format):

````python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(matrix[0, 1])    # 2 (Row 0, Column 1)  
print(matrix[:, 1])    # [2, 5] (All rows, Column 1)
````

##### Boolean Indexing/Masking:

````python
arr = np.array([1, 2, 3, 4])
print(arr[arr > 2])    # [3, 4]
````

````python
arr = np.array([1, 2, 3, 4])
bool_arr = arr > 2
print(bool_arr)
print(arr[bool_arr])    # [3, 4]
````

#### Exercise 1: Statistical Calculations

##### Dataset:

```python
data = np.array([22, 45, 34, 19, 28])  
```

##### Tasks:

- Compute mean, median, and standard deviation.

- Normalize the data to a 0–1 scale.

##### Solutions:


````python
data = np.array([22, 45, 34, 19, 28]) 

mean = np.mean(data)  
median = np.median(data)  
std = np.std(data)  
normalized = (data - np.min(data)) / (np.max(data) - np.min(data))  

print(mean)
print(median)
print(std)
print(normalized)
````

Normalizing data to a 0–1 scale means transforming the values so that they fall within the range [0, 1]. This is a common preprocessing step in data analysis and machine learning to ensure that all features are on a similar scale, which can improve the performance of certain algorithms (e.g., gradient descent in neural networks, distance-based methods like k-NN).

##### Theory Behind Min-Max Normalization (0–1 Scaling)
The most common way to normalize data to a 0–1 scale is using **min-max normalization**. The formula is:

$$
x_{\text{normalized}} = \frac{x - \min(X)}{\max(X) - \min(X)}
$$

Where:
- $x$ = original value
- $\min(X)$ = minimum value in the dataset
- $\max(X)$ = maximum value in the dataset
  
##### Steps to Normalize Data
Given data:  
`data = np.array([22, 45, 34, 19, 28])`

1. **Find the min and max**:
   - $ \min(X) = 19 $
   - $ \max(X) = 45 $

2. **Apply the formula**:
   - For $ x = 22 $:  
     $ \frac{22 - 19}{45 - 19} = \frac{3}{26} \approx 0.115 $
   - For $ x = 45 $:  
     $ \frac{45 - 19}{45 - 19} = \frac{26}{26} = 1.0 $
   - For $ x = 34 $:  
     $ \frac{34 - 19}{45 - 19} = \frac{15}{26} \approx 0.577 $
   - For $ x = 19 $:  
     $ \frac{19 - 19}{45 - 19} = \frac{0}{26} = 0.0 $
   - For $ x = 28 $:  
     $ \frac{28 - 19}{45 - 19} = \frac{9}{26} \approx 0.346 $

3. **Result**:  
   The normalized array is approximately:  
   `[0.115, 1.0, 0.577, 0.0, 0.346]`

##### Key Points
- **Purpose**: Ensures all values are on a consistent scale (0–1).
- **When to Use**: Useful when the absolute scale of the data doesn't matter, but relative differences do (e.g., in neural networks, clustering, or distance-based algorithms).
- **Alternative Normalization**: If outliers are present, other methods like **Z-score standardization** (scaling to mean 0, std 1) may be better.

## **1. Patient Vital Signs Analysis**  
**Dataset:** Heart rate (bpm) and blood pressure (mmHg) for 5 patients.  
```python
import numpy as np
vital_signs = np.array([
    [72, 120, 80],  # Patient 1: HR, SBP, DBP
    [68, 115, 76],
    [88, 145, 90],  # Patient 3 (hypertensive)
    [76, 130, 85],
    [94, 110, 70]   # Patient 5 (tachycardic)
])
```

### **Task:**  
1. Identify Hypertensive Patients (SBP >140 or DBP >90) 
   
2. Calculate Mean Arterial Pressure (MAP)

````python
import numpy as np

# Patient data: [Heart Rate, SBP, DBP]
vital_signs = np.array([
    [72, 120, 80],  
    [68, 115, 76],
    [88, 145, 90],  
    [76, 130, 85],
    [94, 110, 70]   
])

# Extract SBP and DBP
SBP = vital_signs[:, 1]
DBP = vital_signs[:, 2]

# Identify hypertensive patients
hypertensive_mask = (SBP > 140) | (DBP > 90)
hypertensive_patients = vital_signs[hypertensive_mask]

# Calculate Mean Arterial Pressure (MAP)
MAP = (SBP + 2 * DBP) / 3

# Display results
print("Hypertensive Patients (HR, SBP, DBP):")
print(hypertensive_patients)

print("\nMean Arterial Pressure (MAP) for all patients:")
print(MAP)
````

### **Tasks:**  
1. Crop the Image  
   
2. Brighten by 20%

````python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load and convert the image to grayscale
img = Image.open('C:\\Users\\frank\\data_science_project_advance\\datasets\\ui.png').convert('L')
img_array = np.array(img)

# --- Task 1: Crop the image ---
# Define crop boundaries: [start_row:end_row, start_col:end_col]
# Example: crop the center 200x200 region (adjust as needed)
start_row = img_array.shape[0] // 2 - 100
end_row = start_row + 200
start_col = img_array.shape[1] // 2 - 100
end_col = start_col + 200
cropped = img_array[start_row:end_row, start_col:end_col]

# --- Task 2: Brighten the cropped image by 20% ---
brightened = cropped * 1.2
brightened = np.clip(brightened, 0, 255).astype(np.uint8)  # ensure valid pixel range

# --- Plot the original, cropped, and brightened images ---
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(img_array, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(cropped, cmap='gray')
axes[1].set_title("Cropped Image")
axes[1].axis('off')

axes[2].imshow(brightened, cmap='gray')
axes[2].set_title("Brightened Image")
axes[2].axis('off')

plt.tight_layout()
plt.show()
````
