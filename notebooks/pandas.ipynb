#### filtering

````python
# Filter rows where Population < 4000000
filter_df = df[df['Population'] < 4000000]
print(filter_df)
````

````python
# Multiple conditions
filter_df2 = df[(df['Population'] > 25) & (df['City'].str.startswith('K'))]
print(filter_df2)
````
